### kerasлҘј нҷңмҡ©н•ң y = 2x + 1 м„ нҳ•нҡҢк·Җ лӘЁлҚё мҳҲмёЎ мҳҲм ң

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 1. лҚ°мқҙн„° мӨҖл№„
# мһ…л Ҙ(x)кіј м¶ңл Ҙ(y) лҚ°мқҙн„°лҘј м •мқҳн•©лӢҲлӢӨ.
x_data = np.array([0, 1, 2, 3, 4], dtype=np.float32)  # мһ…л Ҙ лҚ°мқҙн„°
y_data = np.array([1, 3, 5, 7, 9], dtype=np.float32)  # м •лӢө лҚ°мқҙн„°
```

```python
# 2. лӘЁлҚё м •мқҳ
# Sequential APIлҘј мӮ¬мҡ©н•ҙ мӢ кІҪл§қмқ„ м •мқҳн•©лӢҲлӢӨ.
model = Sequential([
    Dense(1, input_dim=1)  # мһ…л Ҙ 1к°ң, м¶ңл Ҙ 1к°ң (м„ нҳ• нҡҢк·Җ)
])
```

```python
# 3. лӘЁлҚё м»ҙнҢҢмқј
# мҶҗмӢӨ н•ЁмҲҳмҷҖ мҳөнӢ°л§Ҳмқҙм ҖлҘј м„Өм •н•©лӢҲлӢӨ.
model.compile(optimizer='sgd', loss='mse')
```

```python
# 4. лӘЁлҚё н•ҷмҠө
# fit() л©”м„ңл“ңлЎң лҚ°мқҙн„°лҘј н•ҷмҠөмӢңнӮөлӢҲлӢӨ.
model.fit(x_data, y_data, epochs=50, verbose=1)  # 50лІҲ л°ҳліө н•ҷмҠө
```

```python
# 5. мҳҲмёЎ л°Ҹ нҸүк°Җ
# мғҲлЎңмҡҙ лҚ°мқҙн„°лЎң мҳҲмёЎ мҲҳн–ү
x_new = np.array([5], dtype=np.float32)
y_new = model.predict(x_new)

print(f"x = 5мқј л•Ң, мҳҲмёЎлҗң y = {y_new[0][0]:.2f}")
```
