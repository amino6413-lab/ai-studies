## 활성화 함수

### sigmoid 함수

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
#그래프 그려보기
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### Tanh 함수 (하이퍼볼릭 탄젠트 함수)

```python
import numpy as np
import matplotlib.pylab as plt

def tanh(x):
    return np.tanh(x)

# 그래프 그리기
x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.ylim(-1.1, 1.1)  # y축 범위 설정
plt.title("Tanh Activation Function")
plt.grid()
plt.show()
```

### ReLU 함수

```python
import numpy as np
import matplotlib.pylab as plt

def relu(x):
    return np.maximum(0, x)

# 그래프 그리기
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.5, 5.5)  # y축 범위 설정
plt.title("ReLU Activation Function")
plt.grid()
plt.show()
```
