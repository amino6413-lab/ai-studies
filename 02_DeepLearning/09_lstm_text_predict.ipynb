### мҲ«мһҗмҳҲмёЎ лӘЁлҚё(LSTM)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
# 1. лҚ°мқҙн„° мғқм„ұ (лһңлҚӨ мҲ«мһҗ мӢңнҖҖмҠӨ)
data = np.array([i for i in range(1, 21)])  # 1л¶Җн„° 20к№Ңм§Җмқҳ мҲ«мһҗ
timesteps = 3  # нғҖмһ„мҠӨн…қ кёёмқҙ, sequence lenghtмҷҖ к°ҷмқҖ к°ңл…җ
```

```python
x_data = []
y_data = []
```

```python
# нғҖмһ„мҠӨн…қм—җ л”°лқј мһ…л Ҙ лҚ°мқҙн„°мҷҖ м¶ңл Ҙ лҚ°мқҙн„° мғқм„ұ
for i in range(len(data) - timesteps):
    x_data.append(data[i:i + timesteps])  # мһ…л Ҙ (3к°ңмқҳ мҲ«мһҗ)
    y_data.append(data[i + timesteps])   # м¶ңл Ҙ (лӢӨмқҢ мҲ«мһҗ)

x_data = np.array(x_data).reshape(-1, timesteps, 1)  # (мғҳн”Ң мҲҳ, нғҖмһ„мҠӨн…қ, н”јмІҳ мҲҳ)
y_data = np.array(y_data)  # м¶ңл Ҙк°’
```

```python
# 2. LSTM лӘЁлҚё м •мқҳ
model = Sequential([
    LSTM(10, input_shape=(timesteps, 1)),  # LSTM л Ҳмқҙм–ҙ (мң лӢӣ: 10)
    Dense(1)  # м¶ңл Ҙ л Ҳмқҙм–ҙ
])
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
# 3. лӘЁлҚё н•ҷмҠө
model.fit(x_data, y_data, epochs=1000, verbose=1)  # 100лІҲ л°ҳліө н•ҷмҠө
```

```python
# 4. мҳҲмёЎ н…ҢмҠӨнҠё
test_input = np.array([18, 19, 20]).reshape(1, timesteps, 1)  # мһ…л Ҙ мҳҲмӢң
predicted_value = model.predict(test_input)
```

```python
# кІ°кіј м¶ңл Ҙ
print("мһ…л Ҙ лҚ°мқҙн„°:", test_input.flatten()) # flattenн•ЁмҲҳ : [[[18],[19],[20]]] --> [18,19,20]
print("мҳҲмёЎ к°’:", predicted_value[0][0])
```
